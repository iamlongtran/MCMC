## AR(1) process
$$X_{t+1} = X_t -\alpha (X_{i} - m) + \sigma\xi_{t+1}$$
$\xi_{i} \sim N(0,1)$, independent & identical distributed random variable, iid


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
iter=1e3
alpha=0.1
m=1
sigma=0.5
X0=0
```

```python
X=X0
Xs=[]
for i in range(int(iter)):
	X += -alpha*(X-m)+sigma*np.random.rand()
	Xs.append(X)
```

```python
plt.hist(Xs,100,histtype='step',density=1)
plt.xlabel('X')
plt.ylabel('freq')
plt.show()
plt.plot(range(int(iter)), Xs)
plt.xlabel('iter')
plt.ylabel('X')
plt.show()
```
