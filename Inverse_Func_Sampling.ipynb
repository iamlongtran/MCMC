## Inverse Function Sampling
Given cdf/pdf, the inverse can be evaluated<br>
Indeed, let X be a random variable with cdf F<br>
$$X=F^{-1}(\theta), \,\,\theta \sim U(0,1)$$

```python
import numpy as np
import matplotlib.pyplot as plt
#import scipy.integrate as integrate
from sympy import solve, integrate, Eq
from sympy.abc import x,t
```

```python
def inv_sampling(y):
	#F_inv_X=lambda x: 5*x**4 -> cdf = x^5 -> F^-1=theta^(1/5)
	sol=theta=solve(Eq(integrate(5*x**4,(x,0,t)), y), solver='bisection')
	num_sol=[s.evalf() for s in sol]
	return float(num_sol[0])
```

```python
Theta=np.linspace(0,1,101)
X=[]
for theta in Theta:
	X.append(inv_sampling(theta))
```

```python
plt.plot(Theta,X,'x',label='inv_sampling_numerical')
plt.plot(Theta,Theta**(1/5),label='true_inv_sampling')
plt.xlabel('theta')
plt.ylabel('X')
plt.legend()
```
