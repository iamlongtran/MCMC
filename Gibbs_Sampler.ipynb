## Gibbs Sampler
we wish to sample a random vector $X=(X_1, X_2, ... , X_n)$ with its pdf f(x)<br>
Let $f(x_i | x_j, j \neq i)$ be the conditional pdf of $i^{th}$ component<br>
Gibbs Sampler says to draw $Y_i$ from conditional pdf $f(y_i|x_{j,k})$ for j=0 to n-1, k=2 to n<br>
Rubinstein, Kroese, 2017

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#e.g 6.5 Robeinstein and Kroese
```

$$f(x,y) = c_1(y)\,\, exp(-\dfrac{1+y^2}{2}(x - \dfrac{4}{1+y^2})^2 )$$
derive the normal distribution with its mean and var, we have<br>
$$ X|Y = y \sim N(\dfrac{4}{1+y^2}, \dfrac{1}{1+y^2})\\
Y|X = x \sim N(\dfrac{4}{1+x^2}, \dfrac{1}{1+x^2})
$$



```python
iter=int(1e6)
X=[1]
Y=[1]
```

```python
for i in range(iter):
	x=np.random.normal(4/(1+Y[i]**2), 1/np.sqrt(1+Y[i]**2))
	y=np.random.normal(4/(1+X[i]**2), 1/np.sqrt(1+X[i]**2))
	X.append(x)
	Y.append(y)
```

```python
plt.hist2d(X,Y, bins=(100,100), density=1)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.colorbar()
plt.show()
```

```python
#cumsum
iters=np.array(range(1,iter+1))
X=np.array(X)
cumsum=[]
cumsum=(np.cumsum(X))
cumsum=np.delete(cumsum,0)
cumsum/=iters
plt.plot(iters,cumsum)
plt.xlabel('iter')
plt.ylabel('x_est')
plt.show()
```
